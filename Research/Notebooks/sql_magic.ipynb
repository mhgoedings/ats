# Strategies

```python
%sql select s.id,s.proto_id,s.symbol,s.timeframe_1 tf1,s.timeframe_2 tf2,s.status,s.status_state state,count(t.*) tests \
from strategies s LEFT OUTER JOIN stress_tests t ON (s.id = t.strat_id) \
group by s.id,s.proto_id,s.symbol,tf1,tf2,s.status,state \
having count(t.*) > 0 \
order by s.id
```

```python
%sql select s.id,s.proto_id,s.symbol,s.timeframe_1 tf1,s.timeframe_2 tf2,s.status,s.status_state state,count(t.*) tests \
from strategies s LEFT OUTER JOIN stress_tests t ON (s.id = t.strat_id) \
group by s.id,s.proto_id,s.symbol,tf1,tf2,s.status,state \
having count(t.*) >= 0 \
order by s.id
```
