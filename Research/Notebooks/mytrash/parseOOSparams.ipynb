```python
def metaDataStr(proto_id):
    filepath = ""
    proto_id = ""
    tf1 = ""
    
    str = f'''{{
      Name: {filepath}
      ProtoType run: {proto_id}
      Timeframe 1: {tf1}
      line {3}
}}'''
    return(str)
```

```python
def varBlockStr(vars):
    str = "vars:\n"
    for v in vars.split(','):
        str += "    " + v + "\n"
    return(str)
```

```python
def oosHistoricalParams(inputs,oosStr):
    str = f"// OOS Historical parameters/n"
    for i,l in enumerate(oosStr.rstrip().split('\n')):
        fields = l.split()
        ilist = []
        for inp in inputs.split(','):
            ilist.append(f'{inp} = {fields.pop(0)}')
            #print(f'{inp} = {f.pop(0)}')
        start_dt = fields.pop(0)
        start_yr,start_mn,start_dy = start_dt.split('/')
        _ = fields.pop(0)
        end_dt = fields.pop(0)
        end_yr,end_mn,end_dy = end_dt.split('/')
        str += f"if date >= ELDate({start_mn},{start_dy},{start_yr}) and date < ELDate({end_mn},{end_dy},{end_yr}) then begin\n"
        for p in ilist:
            str += f'    {p}\n'
    return(str)
                  
```
